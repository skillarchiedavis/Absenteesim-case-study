```python
# categories the reasons for absent into groups
reason_1 = reason_columns.loc[:, 1:14].max(axis = 1)
reason_2 = reason_columns.loc[:, 15:17].max(axis = 1)
reason_3 = reason_columns.loc[:, 18:21].max(axis = 1)
reason_4 = reason_columns.loc[:, 22:].max(axis = 1)
```

## Date

```python
type(reasons['Date'][0])
```

```python
# Convert the date column to a datetime format
reasons['Date'] = pd.to_datetime(reasons['Date'])
```

```python
reasons['Date'].head()
```

```python
reasons.info()
```

```python
reasons['Date'][0].month
```

```python
list_months = []
list_months
```

```python
for i in range(reasons.shape[0]):
    list_months.append(reasons['Date'][i].month)
```

```python
list_months[:5]
```

```python
reasons['Month'] = list_months
```

```python
reasons['Month'] = reasons['Month'].map({1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May'
                                        ,6: 'June', 7: 'July', 8: 'August', 9: 'September', 10: 'October', 
                                         11: 'November', 12: 'December'})
reasons['Month'].head()
```

```python
reasons['Date'][699].day_name()
```

```python
Week_day_name, Week_day_numeric = [], []
```

```python
for i in range(reasons['Date'].shape[0]):
    Week_day_numeric.append(reasons['Date'][i].weekday())
    Week_day_name.append(reasons['Date'][i].day_name())
```

```python
reasons['Week_day_numeric'] = Week_day_numeric
reasons['Week_day'] = Week_day_name
```

```python
reasons['Week_day_numeric'].head()
```

```python
reasons['Week_day'].head()
```

```python
Year_list = []
for i in range(reasons['Date'].shape[0]):
    Year_list.append(reasons['Date'][i].year)
```

```python
len(Year_list)
```

```python
reasons['Year'] = Year_list
```

```python
reasons.head()
```

```python
reasons.columns.values
```

```python
column_name =  ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4','Date', 'Year', 'Month', 'Week_day', 'Week_day_numeric',
       'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education', 'Children',
       'Pets', 'Absenteeism Time in Hours']
```

```python
reasons = reasons[column_name]
```
